# Area of residence prediction

```python
house["Area_of_residence"] = 0
house.head()
```

```python
house_information["Area_of_residence"] = 0
house_information.head()
```

```python
# house size (number of bedrooms) available from acorn data
# london houses are smaller than average

house_size = acorn[46:51]
house_size

# 1 bedroom = 40-70 m^2
# 2 bedrooms = 65-92 m^2
# 3 bedrooms = 88â€“120 m^2
# 4 bedrooms = 100-200 m^2
# 5+ bedrooms = 170-300 m^2
```

```python
# calculate the probability that a household is a house or apartment for every ACORN category
acorn = ["ACORN-A", "ACORN-B", "ACORN-C", "ACORN-D", "ACORN-E", "ACORN-F", "ACORN-G", 
         "ACORN-H", "ACORN-I", "ACORN-J", "ACORN-K", "ACORN-L", "ACORN-M", "ACORN-N", 
         "ACORN-O", "ACORN-P", "ACORN-Q"]

def predict_area(acornx):
    if acornx in acorn:
        index = acorn.index(acornx)
    else:
        return 65   # average house size in London
        
    bed = []
    
    # create weighted distribution based on acorn
    for row in range(0,5):
        bed.append(round(house_size.iloc[row,index+3] / house_size[acorn[index]].sum(), 2))

    # sum has to be 1
    while sum(bed) != 1:
        if sum(bed) > 1:
            bed[4] -= 0.01
        else:
            bed[4] += 0.01

    # randomly select amount of bedrooms
    amount_bed = np.random.choice(np.arange(1, 6), p=[bed[0], bed[1], bed[2], bed[3], bed[4]])
    
    # randomly return size of house/apartment based on amount of bedrooms
    if amount_bed == 1:
        return random.randint(40,70)
    elif amount_bed == 2:
        return random.randint(65,92)
    elif amount_bed == 3:
        return random.randint(88,120)
    elif amount_bed == 4:
        return random.randint(100,200)
    else:
        return random.randint(170,300)

# loop every house id in the house_information dataframe and add area_of_residence value
for row in house_information.itertuples():    
    house_information.loc[row.Index, 'Area_of_residence'] = predict_area(row[3])

house_information.head(10)
```
