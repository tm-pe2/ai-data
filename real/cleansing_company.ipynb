```python
import pandas as pd
import numpy as np
```

```python
# Change the path to the correct location of your source data
company = pd.read_csv('./real/Company_data.csv', sep=';');
initial_rows = len(company);
initial_columns = len(company.columns);
# Show initial amount of rows and columns 
print(initial_rows)
print(initial_columns)
# Display the first lines of the dataset
company.head(10)
```

```python
# remove any unnecessary columns (depends on original dataset)
company_reduced = company[['Date','Time','Global_active_power']]
company_reduced.head(10)
```

```python
# remove any 
company_reduced.replace(to_replace=r'^\D+[^\.:/-]?\D*[^\.:/-]?\D*$', value=np.NaN, regex=True, inplace=True)
company_reduced.dropna(inplace=True)
company_reduced = company_reduced[company_reduced['Global_active_power'] > '0']
company_reduced.head(10)
```

```python
# add the necessary columns that will be filled later
company_reduced.insert(2,'Latitude', '0')
company_reduced.insert(3,'Longitude', '0')

company_reduced.insert(4,'Size', '0')
company_reduced.insert(5,'Office_or_factory', '0')
company_reduced.insert(6,'Solarpanels', '0')
company_reduced.insert(7,'Heating_type', '0')
company_reduced.insert(8,'Operating_hours', '0')

company_reduced.head(10)
```

```python
# change path to the location for the desired output
company_reduced.to_csv('./real/company_filtered.csv')
```
