## After looking at these graphs, we can choose which features need to be removed and which need to be added, as well as cleaning and organizing the data.

```python
original_location_path = "C:\De Nayer 2021-2022\Practise Enterprise 2\Commercial data formated - kopie"
regexPaterns = ["^RefBldgLa","^RefBldgMe","^RefBldgSm"]
columnsToKeep = ['Date/Time','Electricity:Facility [kW](Hourly)','Gas:Facility [kW](Hourly)','Water Heater:WaterSystems:Gas [kW](Hourly)']
combined = "(" + ")|(".join(regexPaterns) + ")"
count = 0

for dirname in os.listdir(original_location_path):
    dirPath = original_location_path + "\\" + dirname
    for filename in os.listdir(dirPath):
        
        filePath = dirPath + "\\" + filename
        if (bool(re.search(combined, filename))):
            print(filename)
            company = pd.read_csv(filePath, sep=";")
            company = company[columnsToKeep]
            count = count + (company.isnull().sum())
            company.replace(to_replace=r'^\D+[^\.:/-]?\D*[^\.:/-]?\D*$', value=np.NaN, regex=True, inplace=True)
            company.dropna(inplace=True)
            
            if(bool(re.search("^RefBldg\w+Office", filename))):
                company = company[['Date/Time','Electricity:Facility [kW](Hourly)','Gas:Facility [kW](Hourly)']]
                company.insert(3, 'Total consumption [kW](Hourly)', company['Gas:Facility [kW](Hourly)'] + company['Electricity:Facility [kW](Hourly)'])
                company.insert(4, 'latitude', 0.0)
                company.insert(5, 'Longitude', 0.0)
                company.insert(6, 'is_isolated', 0) # 0 = no, 1 = yes
                company.insert(7, 'has_solarpanels', 0) # 0 = no, 1 = yes
                company.insert(8, 'has_gas_heating', 0) # 0 = no, 1 = yes
                
            elif(bool(re.search("^RefBldg\w+Hotel", filename))):
                company.insert(4, 'Total consumption [kW](Hourly)', company['Gas:Facility [kW](Hourly)'] + company['Electricity:Facility [kW](Hourly)'])
                company.insert(5, 'latitude', 0.0)
                company.insert(6, 'Longitude', 0.0)
                company.insert(7, 'is_isolated', 0) # 0 = no, 1 = yes
                company.insert(8, 'has_solarpanels', 0) # 0 = no, 1 = yes
                company.insert(9, 'has_gas_heating', 0) # 0 = no, 1 = yes
                
            company.to_csv(filePath, sep=";", index=False)
        else:
            if os.path.exists(filePath):
                os.remove(filePath)

                
print(count)      
```
